###### Importing Libraries

```python
import warnings
# import lux
import pandas as pd
import numpy as np
# import seaborn as sns
# import matplotlib.pyplot as plt
# %matplotlib inline
from sklearn.model_selection import train_test_split
# from sklearn.metrics import confusion_matrix
# from sklearn.metrics import accuracy_score
# from sklearn.metrics import classification_report
# from sklearn.metrics import f1_score
from sklearn.utils import resample
# from sklearn.linear_model import LogisticRegression
# from sklearn.feature_selection import chi2
# from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
# from sklearn.ensemble import GradientBoostingRegressor
# from sklearn.ensemble import GradientBoostingClassifier
# from sklearn.svm import SVR
# from sklearn.tree import DecisionTreeRegressor
# from sklearn.ensemble import StackingRegressor
# from sklearn.model_selection import RepeatedKFold
from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_absolute_error
# from sklearn.metrics import mean_squared_error
# import sys
# !{sys.executable} -m pip install xgboost # Made installing XGBoost easier...
import xgboost as xgb
```

###### Rescalling, Resampling, etc.

```python
# Even though not presented, at did another Y-scaled. 
# ----------------------- FIRSTLY I attempted to do a new Y-Scale.
# After running some models and checking by how much it missed by on the texts for younger kids,
# Decided to rescale y to later oversample/downsample it and use it as y.
# Ano 1 and Ano 2 kids are still learning letters and alike, so texts are very similar.
# There is a big jump after that phase ends and then another one after the "1o ciclo" ends.
# Ano 10 does not exist in sample because mandatory texts are theatre plays and poetry.
# And then Ano 11 and 12 are very similar in the past texts recommended for 11 have moved to 12 and vice-versa.
# ----------------------- But LASTLY I attempted to do a new Y-Scale.
# As a conclusion of the project, I realised that for outside PNL predictions, it would be better to have a larger dataset for 3rd to 9th grade instead of a dataset
# that spanned 1st - 12th grade [nobody really write like the books for 1st year-olds, nor like the literature for 12th graders.]

DF = pd.read_excel('Working_DataFrame.xlsx')
def fazer_nova_escala(ano):
    if ano == 1 or ano == 2:
        return 1
    elif ano == 3 or ano == 4:
        return 2
    elif ano == 5 or ano == 6 or ano == 7 or ano == 8 or ano == 9:
        return ano - 2
    elif ano == 10 or ano == 11 or ano == 12:
        return 8

DF["Nova_Escala"] = DF['Ano_Int'].apply(fazer_nova_escala)
warnings.filterwarnings("ignore")
DF["Ano_Esco"] = DF["Ano_Esco"].astype("category")
DF["Idade_Esco"] = DF["Idade_Esco"].astype("category")
DF["Skill_Esco"] = DF["Skill_Esco"].astype("category")
DF["Ciclo_Cat" ] = DF["Ciclo_Int"].astype("category")
warnings.filterwarnings("default")
```

```python
# Eventually I did resampling because I had very unbalanced data, but did it on the new scale.
#|  1| 2| 3|  4| 5|  6| 7| 8|
#| 17|22|76|121|87|108|65|90| -----> mean: 73.5 # Comentado o tratamento que vai ser dado a isto
NE_3 = DF[DF['Nova_Escala'] == 3] ### Not to do anything.
NE_5 = DF[DF['Nova_Escala'] == 5] ### Not to do anything.
NE_7 = DF[DF['Nova_Escala'] == 7] ### Not to do anything.
NE_8 = DF[DF['Nova_Escala'] == 8] ### Not to do anything.
NE_1 = DF[DF['Nova_Escala'] == 1] ### To Oversample
NE_2 = DF[DF['Nova_Escala'] == 2] ### To Oversample
NE_4 = DF[DF['Nova_Escala'] == 4] ### To Downsample
NE_6 = DF[DF['Nova_Escala'] == 6] ### To Downsample
# Mean of "Not to do anything" = 80. n_samples chosen = 80
NE_1_oversampled = resample(NE_1,replace=True,n_samples=80,random_state=15)
NE_2_oversampled = resample(NE_2,replace=True,n_samples=80,random_state=15)
NE_4_downsampled = resample(NE_4,replace=True,n_samples=80,random_state=15)
NE_6_downsampled = resample(NE_6,replace=True,n_samples=80,random_state=15)
DF_resampled = NE_3.append([NE_5,NE_7,NE_8,NE_1_oversampled,NE_2_oversampled,NE_4_downsampled,NE_6_downsampled])
# DF_resampled['Nova_Escala'].value_counts().sort_index().reset_index().rename({'index':'Nova_Escala',0:'#Entradas'}, axis = 1)
```

```python
DF_to_explore = DF_resampled[['Nova_Escala','Ano_Esco','Idade_Esco','Skill_Esco','Ano_Int','Ciclo_Int','Words per sentence',
         'Verbs per Word','Noun per Word','Adj per Word','Other per Word',
          'Syllables per Word','Simp_Word_1 per Word',
          'Poly_Syl per Word','Mono_Syl per Word',
          'Simp_Word_DC per Word','count_SVO_found per word',
          'count_SVO_found_in_a_row per word','count_SVO_elements_in_a_row per word',
          'Uniques_Per_Word','HH_Index']]

x_axis = ['Words per sentence','Verbs per Word',
          'Noun per Word','Adj per Word','Other per Word',
          'Syllables per Word','Simp_Word_1 per Word',
          'Poly_Syl per Word','Mono_Syl per Word',
          'Simp_Word_DC per Word','count_SVO_found per word',
          'count_SVO_found_in_a_row per word','count_SVO_elements_in_a_row per word',
          'Uniques_Per_Word','HH_Index']

x = DF_to_explore[x_axis]

# y_axis will be defined before each of the two models:
# y_axis = ['Ano_Esco'] # y_axis = ['Nova_Escala']

# Others that could be considered:
# y_axis = ['Idade_Esco'] # y_axis = ['Skill_Esco'] # y_axis = ['Ciclo_Int'] 

# By the way, ultimately XGBoost was used, so no scalling or normalisation was done.
```

###### Model Utilimately Used

```python
# For Y --> Ano_Int:
# Grid Searched (Iteratively), see cell below:    
# Best parameters set found on development set:
# {'alpha': 0.1, 'colsample_bytree': 0.8, 'learning_rate': 0.05, 'max_depth': 8, 'n_estimators': 115}

y_axis = ['Ano_Int']
y = DF_to_explore[y_axis]
x_train, x_test, y_train, y_test = train_test_split(x, y,random_state = 40, test_size = 0.2)

xg_reg = xgb.XGBRegressor(objective ='reg:squarederror',
                          colsample_bytree = 0.8,
                          learning_rate = 0.05,
                          max_depth = 8,
                          alpha = 0.1,
                          n_estimators = 115)

# data_dmatrix = xgb.DMatrix(data=x,label=y)
xg_reg.fit(x_train,y_train)
preds = xg_reg.predict(x_test)

print('MAE of Ano_Int Model:',mean_absolute_error(y_test, preds))

real_vs_pred_Ano_Int = pd.concat([y_test.reset_index(),pd.DataFrame(preds)], axis = 1)
real_vs_pred_Ano_Int = real_vs_pred_Ano_Int.rename({'Ano_Int':'Real Ano_Int', 0: 'Predicted Ano_Int'}, axis = 1)

y_axis = ['Nova_Escala']
y = DF_to_explore[y_axis]

x_train, x_test, y_train, y_test = train_test_split(x, y,random_state = 40, test_size = 0.2)

xg_reg = xgb.XGBRegressor(objective ='reg:squarederror',colsample_bytree = 0.8,
                          learning_rate = 0.08,max_depth = 9,
                          alpha = 0,n_estimators = 100)

xg_reg.fit(x_train,y_train)
preds = xg_reg.predict(x_test)
mae = np.sqrt(mean_absolute_error(y_test, preds))

print('MAE of Nova_Escala Model:',mean_absolute_error(y_test, preds))

real_vs_pred_Nova_Escala = pd.concat([y_test.reset_index(),pd.DataFrame(preds)], axis = 1)
real_vs_pred_Nova_Escala = real_vs_pred_Nova_Escala.rename({'Nova_Escala':'Real Nova_Escala', 0: 'Predicted Nova_Escala'}, axis = 1)

real_vs_pred_ys = pd.concat([real_vs_pred_Ano_Int,real_vs_pred_Nova_Escala.drop('index', axis = 1)], axis = 1)
```

```python
from sklearn.model_selection import GridSearchCV

# Set the parameters by cross-validation

tuned_parameters = [{'objective': ['reg:squarederror'],
                     'colsample_bytree': [0.2,0.4,0.5,0.6,0.8],
                     'learning_rate': [0.001,0.01,0.04,0.05,0.06,0.08,0.1],
                     'max_depth': [4,5,6,7,8,9,10], 
                     'alpha': [0,0.1,0.2,0.3,0.6,1,2,5,10]}]

scores = ['neg_mean_absolute_error']

for score in scores:
    print("# Tuning hyper-parameters for %s" % score)
    print()

    clf = GridSearchCV(xgb.XGBRegressor(), tuned_parameters, scoring='neg_mean_absolute_error')
    clf.fit(x_train, y_train)

    print("Best parameters set found on development set:")
    print()
    print(clf.best_params_)
    print()
    print("Grid scores on development set:")
    print()
    means = clf.cv_results_['mean_test_score']
    stds = clf.cv_results_['std_test_score']
    for mean, std, params in zip(means, stds, clf.cv_results_['params']):
        print("%0.3f (+/-%0.03f) for %r"
              % (mean, std * 2, params))
    print()

    print("Detailed classification report:")
    print()
    print("The model is trained on the full development set.")
    print("The scores are computed on the full evaluation set.")
    print()
    y_true, y_pred = y_test, clf.predict(x_test)
    print()
    
# Best parameters set found on development set:    
# MAE: +0.620 (+/-0.095) for {'alpha': 0, 'colsample_bytree': 0.8, 'learning_rate': 0.08, 'max_depth': 9, 'objective': 'reg:squarederror'}
```
